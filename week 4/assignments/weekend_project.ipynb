```python
track_name = ['vampire', 'She Used To Be Mine', 'W.I.T.C.H', 'Hearing Damage', 'Maroon']
artist_name = ['Olivia Rodrigo', 'Sara Bareilles', 'Devon Cole', 'Thom Yorke', 'Taylor Swift']
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
df.info()
```

```python
plt.figure(figsize=(12,7))

sns.barplot(x = df['track'], y = df['energy'], hue= df['track'], legend=False)

plt.title(f'My Top 5 Songs on Spotify \n(over 30 day period)', fontsize=14)

plt.xlabel(f'\nSong Title', fontsize=12)
plt.ylabel(f'Energy Score\n', fontsize=12);
```

According to Spotify, 'energy' is defined as a perceptual measure of intensity and activity. They describe energetic tracks as feeling "fast, loud, and noisy."

```python
plt.figure(figsize=(16,9))

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr,
            square=True,
            cmap='coolwarm',
            annot=True,
            vmax=1,
            vmin=-1,
            mask=mask);
```

```python
sns.heatmap(corr[['energy']].sort_values('energy', ascending=False),
            annot=True,
            vmax=1,
            vmin=-1,
            cmap='coolwarm');
```

Wanting to look closer at the correlations between energy and another audio feature scores, the above heatmap looks at those correlations. Liveness has the strongest correlation with energy based on the top 5 songs played in the last 30 days. Liveness is tied to the presence of an audience during recording. The higher the liveness of a track, the more likely it is to have been performed lived when recorded. This is understandable as live performances feed off the energy of the audience.

Mode is inidicative of whether a song is in a major or minor key. This correlation tells us that when song is minor, the energy scores decrease. 

What is most interesting is that energy and danceability have a negative correlation. When the energy score is increased, the danceability decreases. That does not mean that every song that has high energy is not danceable, but with the songs in the top 5 of the last 30 days, this is true. If we look at Maroon by Taylor Swift, the energy in this song is high which is true. This song moves and has good beat but there is a higher score for speechiness. The higher the speechiness, the more words have both music and speech with Maroon does in the chorus. This song also is in a minor key which can effect the energy of the song as well. These factors influence eachother and create correlations within correlations. Where as some songs with high speechiness (which can be found in rap), may have high energy AND danceability.
